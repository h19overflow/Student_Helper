```python
from langchain_aws import ChatBedrockConverse

llm = ChatBedrockConverse(
    model="global.anthropic.claude-haiku-4-5-20251001-v1:0",
    region_name="ap-southeast-2",
)
```

```python
messages = [
    (
        "system",
        "You are a helpful assistant that translates English to French. Translate the user sentence.",
    ),
    ("human", "I love programming."),
]
for chunk in llm.stream(messages):
    print(chunk)
```

```python
import boto3

region = "ap-southeast-2"
bedrock = boto3.client("bedrock", region_name=region)

resp = bedrock.list_foundation_models()
for m in resp.get("modelSummaries", []):
    print(m["modelId"], "-", m.get("providerName"), "-", m.get("modelName"))
```

```python
from langchain.agents import create_agent
from pydantic import BaseModel, Field
class Translator(BaseModel):
    translation:str = Field(description="The translation of the input sentence into French")
    reason:str = Field(description="The reason for the translation")
agent =  create_agent(
    model=llm,
    tools=[],
    response_format=Translator,

)
```

```python
from langchain_core.prompts import ChatPromptTemplate

prompt = ChatPromptTemplate.from_messages([
    ("system", "You are a helpful assistant that translates English to French. Translate the user sentence."),
    ("human", "{input}"),
])

result =  agent.invoke(prompt.invoke({"input": "I love programming."}))
```
